```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Daily

```python
df1 = pd.read_csv(r"C:\Users\olive\Documents\GitHub\Home-Heating-Analysis\0311859414_day_2018-10-12_to_2018-12-11.csv",
                delimiter=';')
```

```python
df2 = pd.read_csv(r"C:\Users\olive\Documents\GitHub\Home-Heating-Analysis\0311859414_day_2018-12-12_to_2019-02-14.csv",
                delimiter=';')
```

```python
df = pd.concat([df1, df2])
```

```python
insulation_installation_date = '2019-01-02'
```

### Remove dates where I shut off the power

```python
df = df[(df['Date'] > '2018-12-28') | (df['Date'] < '2018-12-21')]
```

```python
df_before = df[df['Date'] < insulation_installation_date]
df_after = df[df['Date'] > insulation_installation_date]
```

```python
before_consumption_group = df_before.groupby(['Average temperature (째C)']).mean()['kWh ']
```

```python
after_consumption_group = df_after.groupby(['Average temperature (째C)']).mean()['kWh ']
```

```python
before_consumption_group.plot(label = 'Before Insulation')
after_consumption_group.plot(label = 'After Insulation')
plt.legend()
plt.show()
```
